```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 산점도 그래프
## dataframe 활용
df = pd.read_excel('../Pandas/score2.xlsx')
# df.set_index('지원번호', inplace=True)
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('수학점수')
plt.ylabel('영어점수')
plt.show()
```

```python
sizes = np.random.rand(8)*1000
sizes
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('수학점수')
plt.ylabel('영어점수')
plt.show()
```

```python
sizes = df['학년']*500
sizes
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('수학점수')
plt.ylabel('영어점수')
plt.show()
```

```python
sizes = df['학년']*500
sizes
plt.scatter(df['영어'], df['수학'], s=sizes,c=df['학년'], cmap='viridis', alpha= 0.5)
plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('수학점수')
plt.ylabel('영어점수')
plt.show()

# google. matplotlib cmap
```

```python
plt.figure(figsize=(7, 7))
sizes = df['학년']*500
sizes
plt.scatter(df['영어'], df['수학'], s=sizes,c=df['학년'], cmap='viridis', alpha= 0.5)
plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('수학점수')
plt.ylabel('영어점수')
plt.colorbar(ticks=[1,2,3], label='학년', shrink=0.5, orientation= 'horizontal')
plt.show()
```
