```python
## CoVid 백신 접종 인구
days = [1,2,3]
az = [2,4,8]
pfizer = [5,1,3]
moderna = [1,2,5]
```

```python
plt.title('Covid 월별 접종건수')
plt.xlabel('월')
plt.ylabel('접종건수(만명)')
plt.xticks([1,2,3])
plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
plt.show()
```

```python
plt.title('Covid 월별 접종건수')
plt.xlabel('월')
plt.ylabel('접종건수(만명)')
plt.xticks([1,2,3])
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label = 'pfizer', marker = 'o', ls = '--')
plt.plot(days, moderna, label = 'moderna', marker = 's', ls = '-.')
plt.legend(ncol=3)
plt.show()
```
