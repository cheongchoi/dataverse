```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
# 스타일
plt.plot(x,y)
plt.title('꺾은선')
```

```python
plt.plot(x, y, linewidth = 5)
```

```python
plt.plot(x,y, marker = 'o')
# marker
```

```python
plt.plot(x,y, marker = 'o', linestyle = 'None')
```

```python
plt.plot(x, y, marker = 'v', markersize=10)
```

```python
plt.plot(x, y, marker = 'X', markersize=10)
```

```python
plt.plot(x, y, marker = 'D', markersize=10)
```

```python
plt.plot(x, y, marker = 'o', markersize=10, markeredgecolor = "red", markerfacecolor = 'yellow')
```

```python
# 선스타일
plt.plot(x, y, linestyle = ":", marker = 'D', markersize=10)
```

```python
plt.plot(x, y, linestyle = "--", marker = 'D', markersize=10)
```

```python
plt.plot(x, y, linestyle = "-.", marker = 'D', markersize=10)
```

```python
# 색깔
plt.plot(x,y, color="pink")
```

```python
# 포맷
plt.plot(x,y, "ro--")
```

```python
plt.plot(x,y, "bv:")
```

```python
plt.plot(x,y, "go")
```

```python
## 축약어
plt.plot(x,y, marker='o', mfc = 'red', ms=10, mec='blue', ls=':')

# matplotlib markeredgecolor 
```

```python
# 투명도
plt.plot(x,y, marker='o', mfc = 'red', ms=10, mec='blue', ls=':', alpha=0.3)
```

```python
# 그래프 크기
plt.figure(figsize = (10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize = (5, 10))
plt.plot(x,y)
```

```python
plt.figure(figsize = (10, 5), dpi = 50)
plt.plot(x,y)
```

```python
## 배경색
plt.figure(facecolor="yellow")
plt.plot(x,y)
```
